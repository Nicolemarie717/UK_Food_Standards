```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20}
result=list(establishments.find(query))

# Use count_documents to display the number of documents in the result
print('The number of the establishments with hygiene score of 20 are:', establishments.count_documents(query))

# Display the first document in the results using pprint
print('The first establishment with hygiene score of 20 is :')
pprint(result[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_db=pd.json_normalize(result)

# Display the number of rows in the DataFrame
print('Number of rows:' ,hygiene_db.shape[0])

# Display the first 10 rows of the DataFrame
hygiene_db.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# ount_documents() method. Hereâ€™s an example that fixes this issue:

filter = {'AddressLine3': 'London',"RatingValue":4,"RatingValue": 5}
cursor = establishments.find(filter)
count = establishments.count_documents(filter)
print('Number of documents:', count)
for document in cursor:
    pprint(document)
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "$and": [
        {"LocalAuthorityName": {'$regex' : "London"}},
        {"$or": [{"RatingValue":4}, {"RatingValue": 5}]}
    ]
}

result = list(establishments.find(query))

# Use count_documents to display the number of documents in the result
print('Number of documents:', establishments.count_documents(query))

# Display the first document in the results using pprint
if result:
    print('First document:')
    print('')
    pprint(result[0])
else:
    print('No documents found')
```

```python
# Convert the result to a Pandas DataFrame
rating=pd.json_normalize(result)

# Display the number of rows in the DataFrame
print('Number of rows are:' , rating.shape[0])

# Display the first 10 rows of the DataFrame
rating.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query = {
    'RatingValue': 5,
    "geocode.longitude": {
        '$gte': longitude - degree_search,
        '$lte': longitude + degree_search
    },
    "geocode.latitude": {
        '$gte': latitude - degree_search,
        '$lte': latitude + degree_search
    }
}

sort = [('scores.Hygiene', 1)]
limit = 5

results = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
for result in results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
top_5=pd.json_normalize(results)
top_5.head(5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene' : 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group' : {'_id': '$LocalAuthorityName', 'count' : {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1 }}

pipeline = [match_query, group_query, sort_values]

result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print('Number of documents in the result:' , len(result))
print('-'*50)
print("")

# Print the first 10 results
pprint(result[0:10])
```

```python
# Convert the result to a Pandas DataFrame
establishments_LocalAuthority=pd.json_normalize(result)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(establishments_LocalAuthority))

# Display the first 10 rows of the DataFrame
establishments_LocalAuthority.head(10)
```
